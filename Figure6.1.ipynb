```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1, 100)

def plot_kernel(f_basis, xd):
    for f in f_basis:
        plt.plot(x, f)
    plt.show()

    idx_xd = np.abs(xd - x).argmin()
    kernel = np.dot(f_basis[:, idx_xd], f_basis)

    plt.plot(x, kernel)
    plt.show()
```

# Power function

```python
num_basis = 10

x = np.linspace(-1, 1, 100)
f_basis = []
for i in range(num_basis):
    f_basis.append(np.power(x, 1 + i))
f_basis = np.array(f_basis)

plot_kernel(f_basis, xd=-0.5)
```

# Gaussian

```python
num_basis = 11

f_basis = []
for i in range(num_basis):
    mu = (-1 + 2. / (num_basis - 1) * i)
    sigma = 0.2
    f_basis.append(np.exp(-np.square((x - mu) / sigma) / 2))
f_basis = np.array(f_basis)

plot_kernel(f_basis, xd=0.0)
```

# Sigmoid

```python
num_basis = 11

f_basis = []
for i in range(num_basis):
    mu = (-1 + 2. / (num_basis - 1) * i)
    sigma = 0.05
    f_basis.append(1. / (1 + np.exp(-(x - mu) / sigma)))
f_basis = np.array(f_basis)

plot_kernel(f_basis, xd=0.0)
```
